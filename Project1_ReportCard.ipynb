# 📋 Project 1 — Student Report Card Generator
> **Concepts:** `list` `tuple` `dictionary` `string` `if-elif-else` `for loop` `function`

---

## 📖 Kya Banana Hai?
Ek system jisme:
- Students ke marks enter karo
- Automatically grade calculate ho
- Sundar report card print ho
- Class topper announce ho

## 🗺️ Code ka Structure
```
setup         → subjects list, grade_scale tuple, students dict
get_grade()   → marks se A+/A/B/C/F return karta hai
add_student() → input leke dictionary mein store karta hai
print_report()→ formatted report card print karta hai
find_topper() → dictionary loop se best student dhundta hai
MAIN          → sab functions ko order mein chalata hai
```

## ▶️ Neeche wala cell Run karo — Poora Project Chalega!
> **Shortcut:** `Shift + Enter`  ya upar se **▶ Run** button dabao

````
# ════════════════════════════════════════════
#   📋 PROJECT 1 — Student Report Card
#   Concepts: list, tuple, dict, if-elif-else,
#             for loop, function, string
# ════════════════════════════════════════════

# ── STEP 1: Data Setup ───────────────────────
subjects    = ["Math", "Science", "English", "History", "Computer"]
grade_scale = (90, 75, 60, 45)   # tuple — fixed, change nahi hoga
students    = {}                  # dictionary — student data store

# ── STEP 2: Grade Function ───────────────────
def get_grade(marks):
    if   marks >= grade_scale[0]: return "A+"
    elif marks >= grade_scale[1]: return "A"
    elif marks >= grade_scale[2]: return "B"
    elif marks >= grade_scale[3]: return "C"
    else:                         return "F"

# ── STEP 3: Add Student ──────────────────────
def add_student(name):
    marks_list = []
    print(f"\n➕ {name} ke marks enter karo:")
    for subject in subjects:
        marks = int(input(f"  {subject}: "))
        marks_list.append(marks)
    students[name] = marks_list
    print(f"✅ {name} ka data save ho gaya!")

# ── STEP 4: Average + Report ─────────────────
def calculate_average(marks_list):
    total = 0
    for m in marks_list:
        total += m
    return total / len(marks_list)

def print_report(name):
    marks_list = students[name]
    avg = calculate_average(marks_list)
    print("\n" + "="*38)
    print(f"   📋 REPORT CARD — {name.upper()}")
    print("="*38)
    for i in range(len(subjects)):
        grade = get_grade(marks_list[i])
        print(f"  {subjects[i]:<12}: {marks_list[i]:>3}   [{grade}]")
    print("-"*38)
    print(f"  {'Average':<12}: {avg:>5.1f}   [{get_grade(avg)}]")
    print("="*38)

# ── STEP 5: Find Topper ──────────────────────
def find_topper():
    top_name, top_avg = "", 0
    for name, marks_list in students.items():
        avg = calculate_average(marks_list)
        if avg > top_avg:
            top_avg, top_name = avg, name
    print(f"\n🏆 Class Topper : {top_name}")
    print(f"   Average      : {top_avg:.1f}  [{get_grade(top_avg)}]")

# ════════════════════════════════════════════
#   🚀 MAIN PROGRAM — Yahan se chalega sab
# ════════════════════════════════════════════
print("="*38)
print("  📚 STUDENT REPORT CARD SYSTEM")
print("="*38)

n = int(input("Kitne students ka data enter karein? "))

for i in range(n):
    name = input(f"\nStudent {i+1} ka naam: ")
    add_student(name)

print("\n📋 Saare Reports:")
for name in students:
    print_report(name)

find_topper()
````
