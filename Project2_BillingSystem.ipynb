# 🏪 Project 2 — Shop Billing System
> **Concepts:** `list` `tuple` `string` `if-elif-else` `for loop` `function`

---

## 📖 Kya Banana Hai?
Ek real dukaan jaisa billing system jisme:
- Menu dikhega sundar table format mein
- Customer items cart mein add karega
- Automatic discount milega (Bronze/Silver/Gold)
- Professional receipt print hogi

## 🗺️ Code ka Structure
```
menu[]            → list of tuples (naam, price, category)
cart[]            → list — customer ke selected items
show_menu()       → for loop se menu table print
add_to_cart()     → validate karke cart mein add
calculate_bill()  → subtotal + discount tiers
print_receipt()   → formatted bill print karta hai
MAIN              → while loop mein order leta hai
```

## 💰 Discount Tiers
| Subtotal | Discount |
|----------|----------|
| Rs. 200+ | 5% (Bronze) |
| Rs. 500+ | 10% (Silver) |
| Rs. 1000+ | 20% (Gold) |

## ▶️ Neeche wala cell Run karo — Poora System Chalega!
> **Shortcut:** `Shift + Enter`  ya upar se **▶ Run** button dabao

````python
# ════════════════════════════════════════════
#   🏪 PROJECT 3 — Shop Billing System
#   Concepts: list, tuple, string,
#             if-elif-else, for loop, function
# ════════════════════════════════════════════

# ── STEP 1: Menu Setup ───────────────────────
# Har product = (naam, price, category)
p1 = ("Burger",     150, "Food")
p2 = ("Pizza",      300, "Food")
p3 = ("Cold Drink",  60, "Drink")
p4 = ("Fries",       80, "Food")
p5 = ("Water",       30, "Drink")
p6 = ("Sandwich",   120, "Food")

menu = [p1, p2, p3, p4, p5, p6]
cart = []

# ── STEP 2: Show Menu ────────────────────────
def show_menu():
    print("\n" + "="*45)
    print("        🍔 ABC FAST FOOD MENU 🍔")
    print("="*45)
    print(f"  {'No.':<5} {'Item':<15} {'Price':>8}  {'Category'}")
    print("-"*45)
    for i, product in enumerate(menu):
        print(f"  {i+1:<5} {product[0]:<15} Rs.{product[1]:>5}  {product[2]}")
    print("="*45)

# ── STEP 3: Add to Cart ──────────────────────
def add_to_cart(choice, quantity):
    if choice < 1 or choice > len(menu):
        print("❌ Galat number!")
        return
    elif quantity <= 0:
        print("❌ Quantity 1 ya zyada honi chahiye!")
        return
    else:
        product = menu[choice - 1]
        cart.append((product[0], product[1], quantity))
        print(f"✅ {product[0]} x{quantity} = Rs.{product[1]*quantity} added!")

# ── STEP 4: Calculate Bill ───────────────────
def calculate_bill():
    if len(cart) == 0:
        print("⚠️  Cart khali hai!")
        return None

    subtotal = 0
    for item in cart:
        subtotal += item[1] * item[2]

    if   subtotal >= 1000: discount, label = 0.20, "Gold   (20% off)"
    elif subtotal >= 500:  discount, label = 0.10, "Silver (10% off)"
    elif subtotal >= 200:  discount, label = 0.05, "Bronze  (5% off)"
    else:                  discount, label = 0.00, "No Discount"

    disc_amount = subtotal * discount
    final_total = subtotal - disc_amount
    return subtotal, disc_amount, final_total, label

# ── STEP 5: Print Receipt ────────────────────
def print_receipt(customer_name):
    result = calculate_bill()
    if result is None:
        return

    subtotal, disc_amount, final_total, label = result

    print("\n" + "="*45)
    print("          🧾 ABC FAST FOOD")
    print("          OFFICIAL RECEIPT")
    print("="*45)
    print(f"  Customer : {customer_name.upper()}")
    print(f"  Items    : {len(cart)}")
    print("-"*45)
    print(f"  {'ITEM':<16} {'QTY':>4} {'PRICE':>7} {'TOTAL':>8}")
    print("-"*45)
    for naam, price, qty in cart:
        print(f"  {naam:<16} {qty:>4} {price:>7} {price*qty:>8}")
    print("-"*45)
    print(f"  {'Subtotal':<34} {subtotal:>8.0f}")
    print(f"  {'Discount ('+label+')':<34} -{disc_amount:>7.0f}")
    print("="*45)
    print(f"  {'💰 TOTAL':<34} Rs.{final_total:>5.0f}")
    print("="*45)
    print("       Shukriya! Phir aana 🙏")
    print("="*45)

# ════════════════════════════════════════════
#   🚀 MAIN PROGRAM
# ════════════════════════════════════════════
print("="*45)
print("   🏪 ABC FAST FOOD — POS SYSTEM")
print("="*45)

customer = input("Customer ka naam: ")
print(f"\nAssalam o Alaikum, {customer}! 👋")

show_menu()

print("\n📝 Order lijiye  |  'done' likhein jab khatam ho")

while True:
    try:
        choice = input("\nItem number (ya 'done'): ")
        if choice.lower() == "done":
            break
        qty = int(input("Quantity: "))
        add_to_cart(int(choice), qty)
    except ValueError:
        print("❌ Sirf number likhein!")

if len(cart) > 0:
    print_receipt(customer)
else:
    print("\nKoi order nahi diya. Allah Hafiz! 👋")
````
